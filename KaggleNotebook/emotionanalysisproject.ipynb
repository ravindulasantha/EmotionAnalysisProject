```python
import pandas as pd
import re  # Import the 're' module for regular expressions

# Read the CSV file into a DataFrame
df = pd.read_csv("/kaggle/input/sentiment-analysis-dataset/sentiment_analysis.csv")

# Define a lambda function to remove URLs using regular expressions
def remove_urls(text):
    return " ".join(re.sub(r'^https?:\/\/.*[\r\n]*', '', x, flags=re.MULTILINE) for x in text.split())

# Apply the lambda function to the 'tweet' column
df["tweet"] = df["tweet"].apply(remove_urls)

# Display the first few rows of the modified DataFrame
print(df.head())
```

```python
import pandas as pd
import string  # Import the 'string' module for punctuation removal

# Read the CSV file into a DataFrame
df = pd.read_csv("/kaggle/input/sentiment-analysis-dataset/sentiment_analysis.csv")

# Define a function to remove punctuations
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

# Apply the function to the 'tweet' column
df["tweet"] = df["tweet"].apply(remove_punctuations)

# Display the first few rows of the modified DataFrame
print(df.head())
```
